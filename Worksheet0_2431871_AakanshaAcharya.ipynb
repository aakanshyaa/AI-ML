Problem 4

Matrix Operations
  <p style="color: #444; margin: 0; font-family: Arial; font-size: 14px;">
    Given: &nbsp;<code>A = [[3,4],[7,8]]</code> &nbsp;and&nbsp; <code>B = [[5,3],[2,1]]</code>
  </p>
</div>

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
print('A =\n', A)
print('B =\n', B)
```

```python
#  Proof 1: A × A⁻¹ = I
A_inv = np.linalg.inv(A)
product = np.round(np.dot(A, A_inv)).astype(int)
print('A⁻¹ =\n', A_inv.round(4))
print('\nA × A⁻¹ =\n', product)
print('\n Equals Identity matrix I')
```

```python
# Proof 2: AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print('AB =\n', AB)
print('\nBA =\n', BA)
print('\nAB == BA?', np.array_equal(AB, BA))
print(' AB ≠ BA  →  Matrix multiplication is NOT commutative')
```

```python
# Proof 3: (AB)ᵀ = BᵀAᵀ
AB_T  = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print('(AB)ᵀ =\n', AB_T)
print('\nBᵀAᵀ =\n', BT_AT)
print('\n(AB)ᵀ == BᵀAᵀ?', np.array_equal(AB_T, BT_AT))
print(' Reversal property of transpose proved')
```

System of Linear Equations


  2x − 3y +  z = −1

   x −  y + 2z = −3
   
  3x +  y −  z =  9


```python
#  Solve AX = B using Inverse Method
coeff = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]], dtype=float)
constants = np.array([-1, -3, 9], dtype=float)

# Method 1: Manual inverse
X_manual = np.dot(np.linalg.inv(coeff), constants)
print('Method 1 — Manual Inverse (X = A⁻¹ · B):')
print(f'  x = {X_manual[0]:.4f}')
print(f'  y = {X_manual[1]:.4f}')
print(f'  z = {X_manual[2]:.4f}')

# Method 2: np.linalg.inv
X_linalg = np.linalg.inv(coeff).dot(constants)
print('\nMethod 2 — np.linalg.inv:')
print(f'  x = {X_linalg[0]:.4f}')
print(f'  y = {X_linalg[1]:.4f}')
print(f'  z = {X_linalg[2]:.4f}')

# Verification
verified = np.allclose(np.dot(coeff, X_linalg), constants)
print(f'\nVerification — A · X ≈ B: {verified} ')
```

10.2  Experiment — How Fast is NumPy?


This experiment benchmarks plain Python lists against NumPy arrays across four operations on 1,000,000-element data (matrices: 1000×1000). All timings are measured in a single session.

```python
# Setup
SIZE = 1_000_000

py_list1 = list(range(SIZE))
py_list2 = list(range(SIZE))
np_arr1  = np.arange(SIZE, dtype=float)
np_arr2  = np.arange(SIZE, dtype=float)

print(f'Data size: {SIZE:,}')
print('Ready to benchmark.')
```

```python
# 1. Element-wise Addition
start = time.time()
result_py = [a + b for a, b in zip(py_list1, py_list2)]
time_py = time.time() - start

start = time.time()
result_np = np_arr1 + np_arr2
time_np = time.time() - start

print('Element-wise Addition (size 1,000,000)')
print(f'  Python : {time_py:.6f} s')
print(f'  NumPy  : {time_np:.6f} s')
print(f'  Speedup: {time_py/time_np:.1f}× faster')
```

```python
# 2. Element-wise Multiplication
start = time.time()
result_py = [a * b for a, b in zip(py_list1, py_list2)]
time_py = time.time() - start

start = time.time()
result_np = np_arr1 * np_arr2
time_np = time.time() - start

print('Element-wise Multiplication (size 1,000,000)')
print(f'  Python : {time_py:.6f} s')
print(f'  NumPy  : {time_np:.6f} s')
print(f'  Speedup: {time_py/time_np:.1f}× faster')
```

```python
# 3. Dot Product
start = time.time()
dot_py = sum(a * b for a, b in zip(py_list1, py_list2))
time_py = time.time() - start

start = time.time()
dot_np = np.dot(np_arr1, np_arr2)
time_np = time.time() - start

print('Dot Product (size 1,000,000)')
print(f'  Python : {time_py:.6f} s')
print(f'  NumPy  : {time_np:.6f} s')
print(f'  Speedup: {time_py/time_np:.1f}× faster')
```

```python
# 4. Matrix Multiplication
# Python: 100×100 (1000×1000 would take several minutes)
M = 100
mat_py1 = [[random.random() for _ in range(M)] for _ in range(M)]
mat_py2 = [[random.random() for _ in range(M)] for _ in range(M)]

start = time.time()
result_py = [[sum(mat_py1[i][k] * mat_py2[k][j] for k in range(M))
              for j in range(M)] for i in range(M)]
time_py_100 = time.time() - start
estimated_1000 = time_py_100 * 1000   # O(n³) scaling

# NumPy: 1000×1000
mat_np1 = np.random.random((1000, 1000))
mat_np2 = np.random.random((1000, 1000))

start = time.time()
result_np = np.dot(mat_np1, mat_np2)
time_np = time.time() - start

print('Matrix Multiplication')
print(f'  Python (100×100)     : {time_py_100:.4f} s')
print(f'  Python (1000×1000)   : ~{estimated_1000:.0f} s  (estimated via O(n³) scaling)')
print(f'  NumPy  (1000×1000)   : {time_np:.4f} s')
print(f'  Estimated Speedup    : ~{estimated_1000/time_np:.0f}× faster')
```

Summary — Why is NumPy Faster?
  
Key Reasons for NumPy's Speed

   Vectorised C-level operations — eliminates Python interpreter loop overhead

   Contiguous memory layout — maximises CPU cache utilisation

   BLAS / LAPACK integration — highly optimised routines for linear algebra

  Multi-threading — matrix operations leverage multiple CPU cores automatically
